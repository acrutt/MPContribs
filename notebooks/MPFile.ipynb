```python
from mpcontribs.io.archieml.mpfile import MPFile
from mpcontribs.io.core.utils import read_csv
```

```python
mpfile_in = MPFile.from_file('mpfile_init.txt') # http://archieml.org
print mpfile_in # see slide #4 for MPFile Components API
```

```python
mpfile = MPFile()
mpfile.add_hierarchical_data({'title': 'Test', 'authors': 'P. Huck, A. Jain'})
print mpfile
```

```python
input_files = mpfile_in.hdata.general['input_files']
df = read_csv(open(input_files['csv']).read())
mpfile.add_data_table('mp-10', df, 'demographics')
print mpfile
mpfile.tdata['mp-10']['demographics'].head()
```

```python
mpfile.add_structure(input_files['cif'], name='Fe3O4')
print mpfile
```
